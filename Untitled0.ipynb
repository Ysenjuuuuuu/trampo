```python
import pandas as pd

# Use the uploaded do drive
file_path = '/content/Cars Datasets 2025.csv'

try:
    df = pd.read_csv(file_path, encoding='latin-1')
except UnicodeDecodeError:
    print("Could not decode using latin-1. Please try a different encoding.")
    # You might want to try 'ISO-8859-1' or other encodings here.
    # As an example, you could try:
    # df = pd.read_csv(file_name, encoding='ISO-8859-1')
    pass # Keep the original code for now if latin-1 fails


print("Visualizando as 5 primeiras linhas do dataset:")
print(df.head())

print("\nInformações sobre o dataset:")
print(df.info())


print("\nAnálise estatística básica:")
print(df.describe())
```

```python
import re

df['Cars Prices'] = df['Cars Prices'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).str.strip()
df['Cars Prices'] = pd.to_numeric(df['Cars Prices'], errors='coerce')

df['HorsePower'] = df['HorsePower'].astype(str).str.replace('hp', '', regex=False).str.replace(' ', '', regex=False)
df['HorsePower'] = df['HorsePower'].apply(lambda x: pd.to_numeric(re.split('[-]', x)[0], errors='coerce') if pd.notna(x) and '-' in str(x) else pd.to_numeric(x, errors='coerce'))

df['Torque'] = df['Torque'].astype(str).str.replace('Nm', '', regex=False).str.replace(' ', '', regex=False)
df['Torque'] = df['Torque'].apply(lambda x: pd.to_numeric(re.split('[-]', x)[0], errors='coerce') if pd.notna(x) and '-' in str(x) else pd.to_numeric(x, errors='coerce'))

df['Total Speed'] = df['Total Speed'].astype(str).str.replace('km/h', '', regex=False).str.strip()
df['Total Speed'] = pd.to_numeric(df['Total Speed'], errors='coerce')


print("\nVerificando os tipos de dados após a limpeza:")
print(df.info())
```

```python
#Média de preços e cavalos de potência
print(f"\nPreço médio dos carros: ${df['Cars Prices'].mean():,.2f}")
print(f"Média de HorsePower: {df['HorsePower'].mean():.2f} hp")

#Conte o número de carros por empresa para encontrar as mais representativas
print("\nNúmero de carros por empresa (Top 10):")
print(df['Company Names'].value_counts().head(10))

#Qual o carro mais caro e mais potente do dataset
carro_mais_caro = df.loc[df['Cars Prices'].idxmax()]
print("\nCarro mais caro:")
print(carro_mais_caro[['Cars Names', 'Company Names', 'Cars Prices']])

carro_mais_potente = df.loc[df['HorsePower'].idxmax()]
print("\nCarro mais potente:")
print(carro_mais_potente[['Cars Names', 'Company Names', 'HorsePower']])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Definir um estilo mais limpo para o gráfico
sns.set_style("whitegrid")

# Contar o número de carros por empresa e pegar as top 10
company_counts = df['Company Names'].value_counts().head(10)

# Criar o gráfico com o Seaborn
plt.figure(figsize=(14, 8))
# Usamos o 'barplot' do seaborn e uma paleta de cores moderna ('viridis')
ax = sns.barplot(x=company_counts.index, y=company_counts.values, palette='viridis')

# Adicionar rótulos de valores em cima de cada barra
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center',
                xytext = (0, 9),
                textcoords = 'offset points')

# Adicionar títulos e rótulos
plt.title('Top 10 Empresas por Número de Carros no Dataset', fontsize=20, fontweight='bold')
plt.xlabel('Empresa', fontsize=14)
plt.ylabel('Número de Carros', fontsize=14)

# Melhorar a visualização dos rótulos do eixo X
plt.xticks(rotation=45, ha='right')

# Remover a moldura superior e direita para um visual mais limpo
sns.despine(top=True, right=True)

plt.tight_layout()
plt.show()
```

```python
# Biblioteca 'seaborn' para visualizações mais bonitas
import seaborn as sns
import matplotlib.pyplot as plt

# Calcular a matriz de correlação das colunas numéricas
numeric_df = df[['Cars Prices', 'HorsePower', 'Total Speed', 'Torque']]
correlation_matrix = numeric_df.corr()

print("Matriz de Correlação entre as variáveis numéricas:")
print(correlation_matrix)

# Visualizar a matriz de correlação no grafico
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mapa de Calor da Matriz de Correlação')
plt.show()
```

```python
# Gráfico de Dispersão relação entre HorsePower e Preço dos carro
plt.figure(figsize=(10, 6))
sns.scatterplot(x='HorsePower', y='Cars Prices', data=df)
plt.title('Relação entre Cavalos de Força e Preço do Carro')
plt.xlabel('Cavalos de Força (hp)')
plt.ylabel('Preço do Carro ($)')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar o estilo para um visual mais limpo
sns.set_style("whitegrid")

plt.figure(figsize=(10, 7))
# O `regplot` adiciona a linha de regressão e a área de confiança
sns.regplot(x='HorsePower', y='Cars Prices', data=df)
plt.title('Correlação Linear entre HorsePower e Preço', fontsize=18, fontweight='bold')
plt.xlabel('Cavalos de Força (hp)', fontsize=14)
plt.ylabel('Preço do Carro ($)', fontsize=14)
plt.grid(True)
plt.show()
```
